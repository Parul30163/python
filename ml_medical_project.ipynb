```python
import string
```

```python
panch= string.punctuation
panch
```

```python
def remove_panch(text):
  mylist =[]
  for i in text.split():
    if i not in panch:
      mylist.append(i)
  return " ".join(mylist)
```

```python
df['Symptoms']=df['Symptoms'].astype(str)
df['Symptoms']=df['Symptoms'].apply(remove_panch)
```

```python
df['Medicine']=df['Medicine'].astype(str)
df['Medicine']=df['Medicine'].apply(remove_panch)
```

```python
df['Causes']=df['Causes'].astype(str)
df['Causes']=df['Causes'].apply(remove_panch)
```

```python

df['Disease']=df['Disease'].astype(str)
df['Disease']=df['Disease'].apply(remove_panch)
```

```python
df['Content'] =df.apply(lambda x : " ".join(x.dropna().astype(str)),axis=1)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df['Content'] = df.apply(lambda x: " ".join(x.dropna().astype(str)), axis=1)
cv= CountVectorizer(max_features=287)
```

```python
metrics= cv.fit_transform(df['Content']).toarray()
```

```python
simalarity =cosine_similarity(metrics)
```
