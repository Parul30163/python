```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
y=df['Sale_Price']
x=scaler.fit_transform(df.drop('Sale_Price',axis=1))
x=pd.DataFrame(x, columns=df.drop(columns=['Sale_Price']).columns) # Pass the data as the first positional argument
x.head()
```

```python
#checking and removing multicollinearity
```

```python
x.corr()
```

```python
k=x.corr()
z=[[str(i),str(j)]for i in k.columns for j in k.columns if (k.loc[i,j]>abs(0.5))& (i!=j)]
z,len(z)
```

Calculating VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_df = x
VIF = pd.Series([variance_inflation_factor(vif_df.values, i) for i in range(vif_df.shape[1])], index=vif_df.columns)
VIF
```

```python
VIF[VIF==VIF.max()].index[0]
```

```python
def MC_remover(data):
    vif = pd.Series([variance_inflation_factor(data.values, i) for i in range(data.shape[1])], index=data.columns)
    if vif.max() > 5:
        print(vif[vif == vif.max()].index[0], 'has been removed') # Indent this line to be inside the 'if' block
        data = data.drop(columns=[vif[vif == vif.max()].index[0]])
        return data
    else: # Indent this line to be at the same level as the 'if'
        print('No Multicollinearity present anymore') # Indent this line to be inside the 'else' block
        return data # Indent this line to be inside the 'else' block
```

```python
# Initialize vif_data with your DataFrame before the loop
vif_data = x  # Replace 'x' with the actual DataFrame you want to process

for i in range(7):
  vif_data = MC_remover(vif_data)

vif_data.head()
```

Remaining Columns

```python
VIF =pd.Series([variance_inflation_factor(vif_data.values, i) for i in range(vif_data.shape[1])], index=vif_data.columns)
VIF , len(vif_data.columns)
```

Train/Test set

```python
x= vif_data
y = df['Sale_Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

Linear Regerssion

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression() # remove the normalize parameter
lr.fit(x_train, y_train)
```

```python
lr.coef_
```

```python
predictions =lr.predict(x_test)
lr.score(x_test, y_test)
```

1. Residuals

```python
residuals =predictions - y_test

residual_table = pd.DataFrame({ 'predictions': predictions, 'Residuals': residuals})
residual_table =residual_table.sort_values(by=['predictions'])
```

```python
z = [i for i in range(int(residual_table['predictions'].max()))]
k = [0 for i in range(int(residual_table['predictions'].max()))]
```

```python
plt.figure(dpi=130 ,figsize=(17,7))
plt.scatter(residual_table['predictions'], residual_table['Residuals'],color ='red',s=2)

plt.plot(z, k, color='red',linewidth =3,label ='regression line')
plt.ylim(-800000,800000) # Changed 'ylims' to 'ylim'
plt.xlabel('Predictions')
plt.ylabel('Residuals')
plt.title('Residuals plot')
plt.legend()
plt.show()
```

```python
#Distribution of Errors
```

```python
plt.figure(dpi=100 ,figsize=(10,7))
plt.hist(residual_table['Residuals'],color ='red',bins=200) # Changed 'bin' to 'bins'
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Distribution of Errors')
plt.show()
```
